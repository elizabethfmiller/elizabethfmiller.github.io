### 2.1 - NaN Values

Let's look for NaN values. We can do this with .isna() to find the NaNs and .sum() to add up the number in each column. 

```python
#checking for NaN values
df.isna().sum()
```

Let's find what percent of the data those are. We can do this by dividing the number of NaN * 100 by the size of the index. We then round the answer to look cleaner. 

```python
#Finding percent NaN
((df.isnull() | df.isna()).sum() * 100 / df.index.size).round(2)
```

We can see that we've got some NaN values. The OTHER_PCT (\~68%) and GRADUATION_RATE (\~70%) column was not included with all school years of data. Since these are likely collinear with the other race columns and the dropout rate, respectively, we can drop those columns entirely. That percentage of NaN value is too big to impute. KDE has a separate set of csvs for graduation rate that we used to find drop out rate in part 1. The graduation rate here is not correct. It is unclear what these numbers even represent. 

To drop we use df.drop and list the columns. 

```python
#Dropping the two mostly NaN columns. 
df = df.drop(['OTHER_PCT', 'GRADUATION_RATE'], axis=1)
df
```

Next, we had some other NaN values in the percent Novice, Apprentice, Proficient and Distinguished columns. Let's see what schools are causing this. We can pull up just the PCT_NOVICE column since if there is a NaN for it, there are NaN values for the rest. 

```python
#Which schools are missing standardized test data?
df[df['PCT_NOVICE'].isna()]
```

Looking through this list of schools, this is where subject knowledge really helps. These are alternative high schools. Alternative schools don't always participate in state testing. Alternative schools are last resorts for students struggling to succeed or are for behavioral issues such as juvenile delinquents. While this could be interesting for our project, we cannot keep them because of the lack of standardized testing data (Novices, Apprentices, etc). So we will drop these rows. We can do this many ways, but one fast way is to create a new df without those rows using .notna().

```python
#Removing NaN rows for KPREP scores
df = df[df['PCT_NOVICE'].notna()]

#Checking they're gone. 
df.info()
```

```python
#Finding percent NaN
((df.isnull() | df.isna()).sum() * 100 / df.index.size).round(2)
```

### 2.2 - Train/Test Split

When running machine learning models, we want the machine to learn on a percent of the data, but we need a way to test it as well... therefore we split the data into 2 chunks. The majority of it will be training the model and a small amount is set aside for testing it. The reason we are splitting so soon is to prevent data leakage. It's very important that the training and testing data never mix. So in order to fill in the NaN values we will need to go ahead and split so that each set can get the NaNs filled. 

To split the data we use the sklearn module train test split. We will split our data into 20% test and 80% training. 

Our target is the drop out rate column (dependent). This will be our y variable. 
The other features will be our X variable (independent).

```python
#Assigning Variables
X = df.drop(columns='DROPOUT_PCT')

#Target
y = df['DROPOUT_PCT']
```

```python
from sklearn.model_selection import train_test_split

#splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Checking it out
X_train.info()
```

```python
X_test.info()
```

### 2.3 - Remove Punctuation

We saw in part 1 that the data has commas in it, dollar signs and percent signs. To clean it let's remove those using regex. This code finds each punctuation and removes them. The inplace=True ensures the changes are kept. 

```python
#Removing commas from train
X_train.replace(',','', regex=True, inplace=True)
#Removing commas from test
X_test.replace(',','', regex=True, inplace=True)
```

```python
#Removing $ from train
X_train.replace('\$','', regex=True, inplace=True)
#Removing $ from test
X_test.replace('\$','', regex=True, inplace=True)
```

```python
#Removing commas from train
X_train.replace('\%','', regex=True, inplace=True)
#Removing commas from test
X_test.replace('\%','', regex=True, inplace=True)
```

In some of the Kids Count data, data points are marked with a star for zeroes, or to represent numbers too minimal to count. So we will replace those with 0's here. 

```python
#Replacing * with NaN train
X_train.replace('\*', "0", inplace=True)

#Replacing * with NaN test
X_test.replace('\*', "0", inplace=True)

#Replacing ** with NaN train
X_train.replace('\**', "0", inplace=True)

#Replacing ** with NaN test
X_test.replace('\**', "0", inplace=True)
```

### 4.2 - Multicollinearity

A big issue with having so many features, is that our columns have collinearity with each other, meaning they are dependent on each other. We want to reduce this because this can lead to unreliable  models.

This is explained better using this excerpt from here: https://pvanb.wordpress.com/2016/04/25/vif-stepwise-variable-selection/

"*Multicollinearity is a phenomenon in which two or more predictor variables in a multiple regression model are highly correlated, meaning that one can be linearly predicted from the others with a substantial degree of accuracy. Issues with multicollinearity are that the standard errors of the affected coefficients tend to be large, the data redundancy in the explanatory variables may result in model overfitting, it may result in erratic changes of coefficient estimates in response to small changes in the model or the data, and if the pattern of multicollinearity in the new data differs from that in the data that was fitted, extrapolation may introduce large errors in the predictions (O’brien 2007; Dormann et al. 2012). Predictive uncertainty caused by multicollinearity thus poses a challenge for predictive environmental niche or species distribution modelling, especially when used to predict the distribution of a species under novel conditions (e.g., future climates or in new areas)*"

Collinearity is a linear pattern between two features (columns).  Multicollinearity is when three or more are linear. In general, an absolute correlation coefficient of >0.7 among two or more predictors indicates the presence of multicollinearity. We saw this in our previous correlation matrix in section 3.3 Correlations. 

So let's try creating a dataset without those highly correlated features. The following code is pretty handy. It checks the correlation of features and adds the name to an empty set if the absolute value of the correlation is greater than 0.6. Then we can drop those columns. We'll do this for train and test, and we can use this shortened dataset to test against our models. 

```python
#Finding correlated features
correlated_features = set()
correlation_matrix = X_train.corr()

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.6:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
```

### 5.2 - K Nearest Neighbors Regression

Linear Regressions have some major draw backs. Mostly due to the way it splits the data only linearly. A lot of data doesn't group well with just straight lines. This is where K Nearest Neighbors comes in. KNN models assign nonlinear boundaries -- they can bend and have jagged edges. 

KKN looks for the nearest data points in distance and their local interpolation to determine a new data point's interpolation. The main drawback of KNN is the amount of time it takes since the algorithm has to predict the distance between ALL of the data points.

To begin, we want to optimize our model by choosing the best number of neighbors. We don't know what that is so we'll run a for loop to find it. And we don't know which scaling technique would be best so we'll test each one. 

By trying to find the best number of neighbors (K), we are setting/finding hyperparameters. This is what the question is referring to with "optimal value". A hyperparameter can't be learned by the model, it's something the user has to set to help the model improve it's performance. 

Again, we're using the rsquared value to compare models against themselves.

We also must scale our data. KNNs are usually very scale sensitive-- because they are calculating distance. If our data points are very spread out, then they tend to suffer. We'll start with standard scaler. 

#### Standard

```python
from sklearn.neighbors import KNeighborsRegressor

# STANDARD SCALER

# The different K values to try out and save to 'neighbors'
neighbors = range(1, 50)

# Empty lists to append to 
train_acc_ss = []
test_acc_ss = []

# Loop through the neighbors 
for k in neighbors: 
    
    # Instantiate the model 
    KNN = KNeighborsRegressor(n_neighbors=k, n_jobs=-1)
    
    # Fit the model 
    KNN.fit(X_train_ss, y_train)
    
    # Score the model 
    train_acc_ss.append(KNN.score(X_train_ss, y_train))
    test_acc_ss.append(KNN.score(X_test_ss, y_test))
```

### 5.3 - Decision Tree Regression

Decision trees are another method of supervised learning similar to KNN. Like KNN, it can handle non-linear decision boundaries, but it's not distance based like KNN. Instead it's like a linear regression... but it does this multiple times so you get straight line boundaries with 90 degree angles.  

Decision trees are similar to flow charts, and can even be turned into one using sklearn. It asks binary questions of each feature and based on the result labels new points as their percentage. This is called hierarchal. 

A real danger of decision trees is that they will overfit and as a result we cannot split it too much. If we avoid default parameters we can reduce this. However, the benefit of DT are that we don't have to use scaled data. 

And, like linear regressions, we can determine the most important features from this in the end. 

Let's run a normal non-optimized decision tree first to see where we stand. 

```python
from sklearn.tree import DecisionTreeRegressor

#Instantiate
my_DT = DecisionTreeRegressor()

#Fit
my_DT = my_DT.fit(X_train, y_train)

#Score
print(f' DT train score: {my_DT.score(X_train, y_train)}')
print(f' DT test score: {my_DT.score(X_test, y_test)}')
```

Here we can see the pitfalls mentioned above. The training data score is much higher than the test (100% vs. 53%!). This is because the data overfit to the training model. The two scores should be close together, like our other models have achieved.  To improve this we will have to set hyperparameters to adjust the max depth (the number of splitting, or levels in the flow chart) the Decision Tree can do.

```python
#Splitting the variables into train/validation/test
from sklearn.model_selection import train_test_split

X_traincv, X_val, y_traincv, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

To find the optimum max depth parameter, we can graph the previous result. We want to see at which depth the Decision Tree began to overfit to the training data. To do this we can use a for loop.

The for loop will check values of 1-10 for the max_depth and compare the rsquared score. Then we will graph the results to see where the train data and test data are closest without overfitting (train data being overly predictive compared to test). We'll use the validation set as the "test" set. 

```python
#For loop to find best max depth (least overfit)

max_depths = np.arange(10)+1

train_scores = []
test_scores = []

for depth in max_depths:
    
    #Instantiate
    my_DT2 = DecisionTreeRegressor(max_depth=depth)
    
    #Fit
    my_DT2.fit(X_traincv, y_traincv)
    
    #Score
    train_scores.append(my_DT2.score(X_traincv, y_traincv))
    test_scores.append(my_DT2.score(X_val, y_val))
```

```python
#Visualizing

plt.figure(figsize=(10,8))
plt.plot(max_depths, train_scores, label="train", marker='.', color='red')
plt.plot(max_depths, test_scores, label="test", marker='.', color='blue')
plt.xticks(np.arange(9)+1)
plt.xlabel("Max Depth")
plt.ylabel("Score")
plt.title("Max_Depth Parameter vs. Score")
plt.legend()
plt.show()
```

Looking at this graph, you can see by the red train line that it veers sharply up past a certain point, and eclipses the blue test line. This is where the model has overfit the data and is less predictive of the test. 

We want to be conservative when dealing with decision trees due to the inclination to overfit. So we're going to pick a Max Depth of 3. 

Now that we know our ideal parameter for max depth, let's rerun the original model with the Max Depth of 3 and see how it performs. We are going to use the exact same code, the only difference is the max_depth = 3 when we instantiate. 

```python
#Instantiate
my_DT = DecisionTreeRegressor(max_depth=4)

#Fit
my_DT = my_DT.fit(X_traincv, y_traincv)

#Score
print(f' DT train score: {my_DT.score(X_traincv, y_traincv)}')
print(f' DT test score: {my_DT.score(X_test, y_test)}')
```

This is an improvement from our first basic DT with a test score of 64% vs. 48%. We can also see that the overfitting is now controlled with how close the test and train are. 

What we are doing here, by preventing the decision tree from gleefully splitting to it's hearts desire, is called "pre-pruning". This is the best method to prevent overfitting. Here, we focused on truncating the max depth of the tree we created. This keeps the model/ tree less complex. 

We can see which columns the Decision Tree is valuing the most: school year and then the white percent, black percent, then by percent novice and attendance rates and then by spending per student, number of suspensions/expulsions, percent on food stamps (SNAP), and then qualifications of teachers.

### 5.4 - Random Forest Regression

Random Forests are an ensemble method of Bagging. Bagging is where we train a bunch of models but each only sees a small portion of the data, then we average them out. Bagging is a mix of bootstrapping and aggregation. Bootstrapping is how we show the models subsets of the data. Aggregation is where we are combining many models and averaging their conclusions. Essentially, to get a random forest, we just run a bunch of Decision Trees on the data and average the results. 

Because Random Forests are Decision Trees, we still get the 90 degree boundary divisions, but with the bagging method, we no longer overfit as badly to the data. And, Random Forests can still be optimized. To start with, let's do a basic Random Forest on the train data. 

```python
from sklearn.ensemble import RandomForestRegressor

#Instantiate
rf = RandomForestRegressor()

#Fit
rf.fit(X_train, y_train)

#Score
print(f'Random Forest Train Score: {rf.score(X_train, y_train)}')
print(f'Random Forest Test Score: {rf.score(X_test, y_test)}')
```

Already we can see that our test score is better than what we've achieved so far at 72%. But we can also see some overfitting since the training score is 95% vs the test score of 72%. So let's start tuning the hyperparameters. We can use a gridsearch to look for them all all at once. This time let's look at more than just max_depth, and look at min_samples_split, min_samples_leaf and n_estimators as well. 

- Max depth refers to how many times a tree can be split. More splits means it captures more data, but it does lead to overfitting (this is fixed by random forests though)
- Min samples split is the minimum number of samples required to split an internal node. These constrain each tree while bagging. 
- Min samples leaf is similar to the previous but refers to how many samples can be at the base of the tree. 
- N estimators is the number of trees that the Random Forest will produce. 

```python
forest = RandomForestRegressor()

n_estimators = [10, 50, 75, 100, 125, 150]
max_depth = [4, 8, 12, 16, 18, 20, 24]
min_samples_split = [2, 4, 6, 8, 10]
min_samples_leaf = [1, 2, 3, 4, 5] 

hyperF = dict(n_estimators = n_estimators, max_depth = max_depth,  
              min_samples_split = min_samples_split, 
             min_samples_leaf = min_samples_leaf)

gridF = GridSearchCV(forest, hyperF, cv = 5, n_jobs = -1)
bestF = gridF.fit(X_train, y_train)
```

```python
bestF.score(X_test, y_test)
```

```python
bestF.best_params_
```

So for example, our regular KNN vs. the hyper optimized one. 

Instead of rsquared value, we could also look at Root Mean Squared Error (RMSE) to determine which model fits the data best. RMSE tells you how spread out the residuals of the regression are. Meaning, how concentrated the data points are around the line of best fit. The lower the RMSE value, the better the model. A score of 0 would be a perfect fit.  

```python
from sklearn.metrics import mean_squared_error

#Linear Regression
y_pred = linreg_act.predict(X_act_test)

rms = mean_squared_error(y_test, y_pred, squared=False)
print(f'Linear Regression RMSE: {rms}')
```

```python
#Decision Tree
y_pred3 = grid_cv_dt.predict(X_corr_test)

rms = mean_squared_error(y_test, y_pred3, squared=False)
print(f'DT RMSE: {rms}')
```
